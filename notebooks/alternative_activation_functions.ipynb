```python
import seaborn as sns
import numpy as np
from scipy.special import expit
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-5, 5, num=100)
```

```python
def relog(x, n=10):
  return np.log(x.clip(min=0)/n + 1) * n
```

```python
data = {
    'ReLU': x.clip(min=0),
    'Sigmoid': expit(x),
    'ReLog n=0.1': relog(x, n=0.1),
    'ReLog n=0.5': relog(x, n=0.5),
    'ReLog n=1': relog(x, n=1),
    'ReLog n=5': relog(x, n=5),
    'ReLog n=10': relog(x, n=10),
}
```

```python
for label, vals in data.items():
    plt.plot(x, vals, label=label)
plt.legend(loc=2)
None
```

The advantage of this compare to $log_n(x + 1/n)+1$ is that it is a generalization of relu. The disadvantage is the transition from relu to relog doesn't have a fixed point. The scale also changes
